## Théorème d’approximation universelle:
    Toute fonction continue f: [a, b] → R peut être approchée d’aussi près que l’on veut par une fonction
    F : [a, b] → R réalisée par un réseau de neurones.
        
    Plusieurs commentaires importants.
    • Tout d’abord rappelons que nous réalisons nos neurones avec des fonctions d’activation H    
      (marche de Heaviside) sauf pour le neurone de sortie qui a pour fonction d’activation l’identité.
    • Les hypothèses sont importantes : f est continue et définie sur un intervalle fermé et borné. 
      Si une des hypothèses venait à manquer l’énoncé serait faux.
    • Bien que l’énoncé ne le dise pas, on peut concrètement réaliser à la main un réseau qui approche 
      la fonction f (voir le chapitre suivant). Cependant, ce n’est pas l’esprit des réseaux de neurones.
    • Que signifie « approcher d’aussi près que l’on veut la fonction f » ?
        
    C’est dire que pour chaque ε > 0, il existe une fonction F (ici issue d’un réseau de neurones) telle que:
    - pour tout x ∈ [a, b] |f (x) − F(x)| < ε

<font face="sans-serif" color=red size=6ptx style="line-height:35px;">
<strong>Gradient</strong></font><br>
<font face="sans-serif" color="#00000" size=4ptx style="line-height:25px;">
Le gradient est un vecteur qui remplace la notion de dérivée pour les fonctions de plusieurs variables.<br> On sait que la dérivée permet de décider si une fonction est croissante ou décroissante.<br>    
De même le vecteur gradient indique la direction vers laquelle la fonction croît ou décroît le plus vite.<br>Nous allons calculer de façon algorithmique le gradient grâce à la « différentiation automatique ».<br> 
<font face="sans-serif" color="#0000FF" size=5ptx style="line-height:25px;">
Dérivées partielles</font><br>
<font face="sans-serif" color="#00000" size=4ptx style="line-height:25px;">
Pour une fonction de plusieurs variables, il existe une dérivée par variable appellée dérivée partielle.<br>
Soit $f:R^2→R$. $\displaystyle \frac{\partial f}{\partial x}(x_0, y_0)$, dérivée partielle de $f$ par rapport à $x$ au point $(x_0, y_0) ∈ R^2$ tel que :
    
 - $\displaystyle \frac{\partial f}{\partial x}(x_0, y_0)$, est la dérivée en $x_0$ de la fonction d'une variable $x→f(x,y_0)$.
 - $\displaystyle \frac{\partial f}{\partial y}(x_0, y_0)$, est la dérivée en $y_0$ de la fonction d'une variable $x→f(x_0,y)$.
    
Autrement dit en revenant à la définition de la dérivée comme une limite : 
  
$ \displaystyle \frac{\partial f}{\partial x}(x_0, y_0) = \lim\limits_{\substack{h \rightarrow 0 }}\displaystyle \frac{f(x_0+h,y_0)-f(x_0,y_0)}{h}$ et $ \displaystyle \frac{\partial f}{\partial y}(x_0, y_0) = \lim\limits_{\substack{h \rightarrow 0 }}\displaystyle \frac{f(x_0,y_0+h)-f(x_0,y_0)}{h}$<br>

Plus généralement, pour une fonction $f : R_n → R$ de plusieurs variables: <br> 
 - $\displaystyle \frac{\partial f}{\partial x}(x_1,...,x_n)$ est la dérivée partielle de $f$ par rapport à la variable $x_i$ au point $(x_1,...,x_n) ∈ R_n$.    

C’est la dérivée en $x_i$ de la fonction d’une variable $x_i→f(x_1,..,x_n)$ considèrant fixes les variables $x_j$ pour $j \neq i$





</font>

<font face="sans-serif" color="#0000FF" size=6ptx style="line-height:35px;">
<strong>Interprétation géométrique</strong></font><br>
<font face="sans-serif" color="#00000" size=4ptx style="line-height:25px;">
Pour une fonction d’une variable, la dérivée est la pente de la tangente au graphe de la fonction (le graphe
étant alors une courbe). Pour une fonction de deux variables $(x, y) → f (x, y)$, les dérivées partielles
indiquent les pentes au graphe de $f$ selon certaines directions (le graphe étant ici une surface).<br>
    
Plus précisément :

 - $ \displaystyle \frac{\partial f}{\partial x}(x_0, y_0)$ est la pente du graphe de $f$ en $(x_0, y_0)$ suivant l’axe $0x$, pente de la tangente à la courbe $z = f (x, y_0)$ donnée par la dérivée $x → f (x, y_0)$ en $x_0$, qui est donc bien $ \displaystyle \frac{\partial f}{\partial x}(x_0, y_0)$•   

 - $ \displaystyle \frac{\partial f}{\partial y}(x_0, y_0)$ est la pente du graphe de $f$ en $(x_0, y_0)$ suivant l’axe $0y$ pente de la tangente à la courbe $z = f (x_0, y)$ donnée par la dérivée $y → f (x_0, y)$ en $y_0$, qui est donc bien $ \displaystyle \frac{\partial f}{\partial y}(x_0, y_0)$•  

Figure de gauche, la dérivée partielle $ \displaystyle \frac{\partial f}{\partial x}(x_0, y_0)$ indique la pente de la tranche parallèle à l’axe $(Ox$) (en orange).<br>
Figure de droite, la dérivée partielle $ \displaystyle \frac{\partial f}{\partial y}(x_0, y_0)$indique la pente de la tranche parallèle à l’axe $(Ox)$ (en vert).
</font>

<font face="sans-serif" color=blue size=5ptx style="line-height:35px;">
<strong>Gradient</strong></font><br>
<font face="sans-serif" color=black size=4ptx style="line-height:25px;">
Le gradient est un vecteur dont les coordonnées sont les dérivées partielles. Il a de nombreuses applications
géométriques car il donne l’équation des tangences aux courbes et surfaces de niveau. Surtout, il indique la
direction dans laquelle la fonction varie le plus vite.
    
Soit $f:R^2→R$ admet les dérivées partielles le gradient de $f$ en $(x_0, y_0)$ ∈ $R^2$ noté $grad f (x_0,y_0)$, est le vecteur:
    
<br>    
<center>
$grad f (x_0,y_0) =  
\begin{pmatrix}
   \displaystyle \frac{\partial f}{\partial x}(x_0, y_0)  \\
   \displaystyle \frac{\partial f}{\partial y}(x_0, y_0)
\end{pmatrix}$
</center>
<br>    
    Les physiciens et les anglo-saxons notent souvent $\nabla f (x, y)$ pour $grad f(x, y)$. Le symbole $\nabla$ se lit « nabla ».<br>   
Plus généralement, pour $f : R^n → R$, le gradient de $f$ en $(x_1,. .. , x_n)$ ∈ $R^n$ est le vecteur :<br>
<br>    
<center>
$grad f (x_1,...,x_n) =  
\begin{pmatrix}
   \displaystyle \frac{\partial f}{\partial x_1}(x_1,...,x_n)  \\
   . \\
   \displaystyle \frac{\partial f}{\partial x_n}(x_1,...,x_n)
\end{pmatrix}$
</center>
<br> 
<font face="sans-serif" color=blue size=4ptx style="line-height:25px;">
<i>Remarque : <br>
Le gradient est un élément de $R^n$ écrit comme un vecteur colonne, parfois, pour alléger l’écriture, on peut également l’écrire sous la forme d’un vecteur ligne.</i>   
</font>
</font>


<font face="sans-serif" color=red size=5ptx style="line-height:35px;">
<strong>Tangentes aux lignes de niveau : </strong></font><br>
<font face="sans-serif" color=black size=4ptx style="line-height:25px;">
Soit $f : R^2 → R$ une fonction différentiable. On considère les lignes de niveau $f (x, y) = k$.<br><br>
<strong>Proposition 1 :</strong> Le vecteur gradient grad $f (x_0,y_0)$ est orthogonal à la ligne de niveau de $f$ passant au point $(x_0, y_0)$<br><br>
Sur ce premier dessin, sont dessinés la ligne de niveau passant par le point $(x_0, y_0)$ (en rouge), un vecteur tangent $v$ en ce point et la tangente à la ligne de niveau (en vert).<br><br>
Le vecteur gradient est un vecteur du plan qui est orthogonal à la ligne de niveau en ce point (en bleu).
<div> 
    <img src="image-10.png" style="width:400px;"/>
</div><br>
À chaque point du plan, on peut associer un vecteur gradient. Ce vecteur gradient est orthogonal à la ligne
de niveau passant par ce point. Nous verrons juste après comment savoir s’il est orienté « vers le haut » ou
« vers le bas ».<br>
Dans le cadre de notre étude, nous nous intéressons à l’équation de la tangente.<br><br>
<strong>Proposition 2 :</strong> Au point $(x_0, y_0)$, l’équation de la tangente à la ligne de niveau de $f$ est : <br><br>
<center>
$ \displaystyle \frac{\partial f}{\partial x}(x_0, y_0)(x - x_0) { + } \frac{\partial f}{\partial y}(x_0, y_0)(y - y_0) = 0$    
</center><br> 
<strong>Exemple (Tangentes à une ellipse).</strong><br>
Trouver les tangentes à l’ellipse E d’équation $\displaystyle \frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$.
<div> 
    <img src="image-11.png" style="width:400px;"/>
</div><br>
Cette ellipse E est la ligne de niveau $f (x, y) = 1$ de la fonction $f (x, y) = \displaystyle \frac{x^2}{a^2} + \frac{y^2}{b^2} $.<br> 
Les dérivées partielles en $(x0, y0)$ sont :
$ \displaystyle \frac{\partial f}{\partial x}(x_0, y_0) = \frac{2x_0}{a^2} $ et $ \displaystyle\frac{\partial f}{\partial y}(x_0, y_0)= \frac{2y_0}{b^2}$   

L’équation de la tangente à l’ellipse E en ce point est donc :
$\displaystyle \frac{2x_0}{a^2}*{(x − x_0)}$ + $\displaystyle \frac{2y_0}{b^2}*{(y − y_0)}{=0}$<br>
Mais comme $\displaystyle \frac{x_0^2}{a^2} + \frac{y_0^2}{b^2} = 1$ <br> 
l’équation de la tangente se simplifie en $\displaystyle \frac{x_0^2}{a^2}*{x} + \frac{y_0^2}{b^2}*{y} = 1$<br>
<font face="sans-serif" color=red size=5ptx style="line-height:35px;">
<strong>Lignes de plus forte pente : </strong></font><br>
Considérons les lignes de niveau $f (x, y) = k$ d’une fonction $f : R2 → R$. On se place en un point $(x0, y0)$. On cherche dans quelle direction se déplacer pour augmenter au plus vite la valeur de $f$.<br>
<br><strong>Proposition 3</strong><br>
Le vecteur gradient $grad f (x_0, y_0)$ indique la direction de plus grande pente à partir du point $(x_0, y_0)$. Autrement dit, si l’on veut, à partir d’un point donné $(x_0, y_0)$ de niveau $a$, passer au niveau $b > a$ le plus vite possible alors il faut démarrer en suivant la direction du gradient $grad f (x_0, y_0)$
<div> 
    <img src="image-12.png" style="width:400px;"/>
</div><br>
Comme illustration, un skieur de descente, voulant optimiser sa course, choisira en permanence de s’orienter
suivant la plus forte pente, c’est-à-dire dans le sens opposé au gradient.<br>
<font face="sans-serif" color=red size=5ptx style="line-height:35px;">
<br><strong>Dérivée directionnelle : </strong></font><br>
Pour prouver que le gradient indique la ligne de la plus grande pente, nous avons besoin de généraliser la
notion de dérivée partielle. Ce passage est plus technique et peut être ignoré en première lecture.<br>
Soit $v = \begin{pmatrix} {h}\\[3mm]{k}\end{pmatrix}$ un vecteur du plan, la dérivée directionnelle de $f$ suivant $v$ en $(x_0, y_0)$ est le nombre :
<center>    
$ \displaystyle D_vf(x_0,y_0) = h * \frac{\partial f}{\partial x}(x_0, y_0)  +  k * \frac{\partial f}{\partial y}(x_0, y_0)$
</center>    
<br>La dérivée directionnelle correspond à la pente de la fonction pour la tranche dirigée par le vecteur $v$.
<div> 
    <img src="image-13.png" style="width:400px;"/>
</div><br>
</font>

<font face="sans-serif" color=black size=4ptx style="line-height:25px;">
    <strong>Remarque :</strong><br><br>
pour $v = \begin{pmatrix} {1}\\[3mm]{0}\end{pmatrix}$  alors $ \displaystyle D_vf(x_0,y_0) = \frac{\partial f}{\partial x}(x_0, y_0) $ et pour $v = \begin{pmatrix} {0}\\[3mm]{1}\end{pmatrix}$ alors $ \displaystyle D_vf(x_0,y_0) = \frac{\partial f}{\partial y}(x_0, y_0)$<br>
On rappelle que le produit scalaire de deux vecteurs $u = \begin{pmatrix} {x}\\[3mm]{y}\end{pmatrix}$ et $v = \begin{pmatrix} {x'}\\[3mm]{y'}\end{pmatrix}$ est donné :<br> 
 - algébriquement par :  $〈u | v〉 = xx′ + yy′$.<br>
 - géométriquement par : $(u | v〉 = ∥u∥ * ∥v∥ · cos(θ)$ (où θ est l’angle entre u et v).
<div> 
    <img src="image-14.png" style="width:400px;"/>
</div><br>
Ainsi, on peut réécrire la dérivée directionnelle sous la forme : <br>   
<font face="sans-serif" color=blue size=4ptx style="line-height:35px;">
<strong>
<center>$\displaystyle D_vf(x_0,y_0) = \nabla f\langle (x_0, y_0) | v\rangle $</center>
</strong>
</font> 
On peut maintenant prouver que le gradient indique la ligne de plus grande pente.<br>
<strong>Démonstration.</strong><br><br> 
La dérivée suivant le vecteur non nul $v$ au point $(x_0, y_0)$ décrit la variation de $f$ autour de
ce point lorsqu’on se déplace dans la direction $v$. La direction selon laquelle la croissance est la plus grande est celle du gradient de $f$. <br>
<font face="sans-serif" color=blue size=4ptx style="line-height:35px;">
<strong>
<center>$\displaystyle D_vf(x_0,y_0) = \nabla f\langle (x_0, y_0) | v\rangle  =\|grad f(x_0, y_0)\|*\|v\|*cos\theta$</center>
</strong>
</font> 
où $\theta$ est l’angle entre le vecteur $grad f (x_0, y_0)$ et le vecteur $v$. <br>
Le maximum est atteint lorsque l’angle $\theta = 0$, c-à-d lorsque $v$ pointe dans la même direction que $grad f (x_0, y_0)$.
</font> 


<font face="sans-serif" color=red size=5ptx style="line-height:25px;">
<strong>Surface de niveau :</strong><br>
<font face="sans-serif" color=black size=4ptx style="line-height:25px;">
Les résultats présentés ci-dessus pour les fonctions de deux variables se généralisent aux fonctions de trois variables ou plus. Commençons avec trois variables et une fonction $f : R3 → R$.<br> 
Rappelons qu’un plan de $R^3$ passant par $(x_0, y_0, z_0)$ et de vecteur normal $n = (a, b, c)$ a pour équation cartésienne: <br>
<font face="sans-serif" color=blue size=4ptx style="line-height:25px;"><center><strong>$a(x − x_0) + b(y − y_0) + c(z − z_0) = 0$</strong></center>
<font face="sans-serif" color=black size=4ptx style="line-height:25px;">
De même qu’il existe une droite tangente pour les lignes de niveau, il existe un <font face="sans-serif" color=orange size=4ptx style="line-height:25px;">plan tangent </font>à une surface de niveau.    
<strong>Prosposition 4</strong> <br> 
Le vecteur gradient $grad f(x_0,y_0,z_0)$ est orthogonal à la surface de niveau de $f$ passant au point $(x_0,y_0,z_0)$.
Autrement dit, l’équation du plan tangent à la surface de niveau de $f$ en $(x_0,y_0,z_0)$ est : 
<center>
<font face="sans-serif" color=blue size=4ptx style="line-height:25px;">    
$ \displaystyle \frac{\partial f}{\partial x}(x_0, y_0, z_0)(x - x_0) { + } \frac{\partial f}{\partial y}(x_0, y_0, z_0)(y - y_0){ + } \frac{\partial f}{\partial z}(x_0, y_0, z_0)(z - z_0) = 0$    
</font>
</center><br> 
pourvu que le gradient de $f$ en ce point ne soit pas le vecteur nul.
<div> 
    <img src="image-15.png" style="width:500px;"/>
</div><br>
Plus généralement pour $f : Rn → R, grad f (x_1,. .. , x_n)$ est orthogonal à l’espace tangent à l’hypersurface de niveau f = k passant par le point $(x_1,. .. , x_n) ∈ R^n$ et ce vecteur gradient $grad f (x_1,. .. , x_n)$ indique la direction de plus grande pente à partir du point $(x_1,. .. , x_n)$.    
</font>

<font face="sans-serif" color=red size=5ptx style="line-height:25px;">
<strong>Calcul approché:</strong><br>
<font face="sans-serif" color=black size=4ptx style="line-height:25px;">

sans calculatrice. <br>
Voici, en deux variables, l’analogue de la formule pour une variable, cette approximation est valable pour $h$ et $k$ petits :<br>
<center>
<br>
$\displaystyle f (x_0 + h, y_0 + k) \approx f (x_0, y_0) + h\frac{\partial f}{\partial x}(x_0, y_0) + k\frac{\partial f}{\partial y}(x_0, y_0)$
<br>
</center> 

Sur l’interprétation géométrique on approche le graphe de $f$ en $(x_0, y_0)$ par le plan tangent au graphe en ce point. <br>
Sur la figure sont représentés le graphe de $f$ (en rouge), le plan tangent au-dessus du point $(x_0, y_0)$ (en bleu). <br><br>
L'exacte valeur $z_1=f(x_0+h, y_0+k)$ par le point de la surface au dessus de $(x_0+h,y_0+k)$. <br>
On approche $\displaystyle z_2 = f (x_0, y_0) + h\frac{\partial f}{\partial x}(x_0, y_0) + k\frac{\partial f}{\partial y}(x_0, y_0)$ par le point du plan tangent dessus $(x_0+h,y_0+k)$

<font face="sans-serif" color=red size=5ptx style="line-height:25px;">
<strong>Minimum et maximum :</strong><br>
<font face="sans-serif" color=black size=4ptx style="line-height:25px;">
<strong>Définition :</strong><br>
Soit $f : R^2 → R$<br>
$f$ admet <font face="sans-serif" color=orange size=4ptx style="line-height:25px;">un minimum local</font> en $(x_0, y_0) \Longleftrightarrow \exists$ un disque $D$ centré en ce point /
$f (x, y) \geqslant f (x_0, y_0)$ $\forall$ $(x, y) \in D$<br>
$f$ admet <font face="sans-serif" color=orange size=4ptx style="line-height:25px;">un maximum local</font> en $(x_0, y_0)$ pour lequel $f(x, y)\leqslant f(x_0,y_0)$ $\forall$  $(x, y) \in D$.<br>
On parle d’<font face="sans-serif" color=orange size=4ptx style="line-height:25px;">extremum local</font> pour un minimum ou un maximum local<br>
<strong>Exemple :</strong><br>
L’exemple type de minimum est celui de la fonction $f (x, y) = x^2 + y^2$ en $(0,0)$. <br>
Voici son graphe et ses lignes de niveau.
<div> 
    <img src="image-17.png" style="width:300px;"/>
</div><br> 
La fonction $f (x, y) = −x^2 − y^2$ admet, elle, un maximum en $(0,0)$.
<div> 
    <img src="image-18.png" style="width:300px;"/>
</div><br> 
 </font>

<font face="sans-serif" color=black size=4ptx style="line-height:25px;">
<strong>Proposition </strong><br>
Soit $f : R2 → R$. Si $f$ admet un minimum ou un maximum local en $(x_0, y_0)$ alors le gradient est le vecteur nul en ce point, autrement dit :<br>
<center>
$\displaystyle \frac{\partial f}{\partial x}(x_0, y_0) = 0$   et   $\displaystyle \frac{\partial f}{\partial y}(x_0, y_0)=0$
</center>

<strong>Démonstration : </strong><br> 
Prenons le cas d’un minimum local.<br>
La fonction qui a $x→f(x, y_0)$ admet un minimum en $x_0$ donc sa dérivée est nulle en $x_0$, c-à-d $\displaystyle \frac{\partial f}{\partial x}(x_0, y_0)=0$.<br>
De même pour $y→f(x_0, y)$ admet un minimum en $y_0$ donc : $\displaystyle \frac{\partial f}{\partial y}(x_0, y_0)=0$<br>
Dans la suite du cours nous chercherons les points pour lesquels une fonction donnée présente un minimum
local. D’après la proposition précédente, ces points sont à chercher parmi les points en lesquels le gradient s’annule. <br>
  On dira que $(x_0, y_0)$ est un point critique de $f$ si les dérivées partielles $\displaystyle \frac{\partial f}{\partial x}$ et $\displaystyle \frac{\partial f}{\partial y}$, s'annulent simultanément.
 <strong>Exemple :</strong>
Chercher les points en lesquels $f(x, y) = x^2 − y^3 + xy$ peut atteindre son minimum. <br>
Recherche des points critiques. <br>
On calcule $ \displaystyle \frac{\partial f}{\partial x}{(x, y)} = 2x + y$ et $ \displaystyle \frac{\partial f}{\partial y}{(x, y)} = −3y^2 + x$.<br>
On cherche les points $(x, y)$ en lesquels les deux dérivées partielles s’annulent. <br>
Par l’annulation de la première dérivée, on a $2x + y = 0$ donc $y = −2x$. <br>
Par l’annulation de la seconde dérivée, on a $−3y^2+x = 0$<br>
Ceci donne par substitution $−12*x^2 + x = 0$, ainsi $x(−12*x + 1) = 0$. <br>
Donc soit $x = 0$ et alors on a $y = 0$, soit $ \displaystyle x =\frac{1}{12}$ et alors $ \displaystyle y =-\frac{1}{6}$.<br>
<strong>Bilan :</strong> 
Il y a deux points critiques :$(0, 0)$ et $ \displaystyle (\frac{1}{12}$,$ -\displaystyle \frac{1}{6})$.<br>
<strong>Étude du point critique $(0,0)$.</strong><br>
On a $f (0,0) = 0$ mais on remarque que $f (0, y) = −y^3$ peut être négatif ou positif (selon le signe de $y$ proche de 0), donc en $(0,0)$ il n’y a ni minimum ni maximum.<br><br>
<strong>Étude du point critique $ \displaystyle (\frac{1}{12}, -\displaystyle \frac{1}{6})$</strong> <br>
Il existe un critère (que l’on ne décrira pas ici) qui permet de dire qu’en ce point $f$ admet un minimum local.
Sur le dessin ci-dessous, le minimum est situé à l’intérieur du petit ovale, l’autre point critique en $(0, 0)$
correspond à l’intersection de la ligne de niveau $f = 0$ avec elle-même.

à deux inconnues.<br>
Il faut prendre garde que ce n’est pas un système linéaire et que dans le cas d’une fonction plus compliquée il aurait été impossible de déterminer exactement les points critiques.<br><br>
On note aussi dans l’exemple précédent que certains points critiques ne sont ni des maximums ni des
minimums.<br>
L’exemple type illustré ci-dessous, est celui d’un col appelé point-selle en référence à sa forme de selle de cheval.<br><br>
