## ThÃ©orÃ¨me dâapproximation universelle:
    Toute fonction continue f: [a, b] â R peut Ãªtre approchÃ©e dâaussi prÃ¨s que lâon veut par une fonction
    F : [a, b] â R rÃ©alisÃ©e par un rÃ©seau de neurones.
        
    Plusieurs commentaires importants.
    â¢ Tout dâabord rappelons que nous rÃ©alisons nos neurones avec des fonctions dâactivation H    
      (marche de Heaviside) sauf pour le neurone de sortie qui a pour fonction dâactivation lâidentitÃ©.
    â¢ Les hypothÃ¨ses sont importantes : f est continue et dÃ©finie sur un intervalle fermÃ© et bornÃ©. 
      Si une des hypothÃ¨ses venait Ã  manquer lâÃ©noncÃ© serait faux.
    â¢ Bien que lâÃ©noncÃ© ne le dise pas, on peut concrÃ¨tement rÃ©aliser Ã  la main un rÃ©seau qui approche 
      la fonction f (voir le chapitre suivant). Cependant, ce nâest pas lâesprit des rÃ©seaux de neurones.
    â¢ Que signifie Â« approcher dâaussi prÃ¨s que lâon veut la fonction f Â» ?
        
    Câest dire que pour chaque Îµ > 0, il existe une fonction F (ici issue dâun rÃ©seau de neurones) telle que:
    - pour tout x â [a, b] |f (x) â F(x)| < Îµ

<strong>Gradient</strong></font><br>
<font face="sans-serif" color="#00000" size=4ptx style="line-height:25px;">
Le gradient est un vecteur qui remplace la notion de dÃ©rivÃ©e pour les fonctions de plusieurs variables.<br> On sait que la dÃ©rivÃ©e permet de dÃ©cider si une fonction est croissante ou dÃ©croissante.<br>    
De mÃªme le vecteur gradient indique la direction vers laquelle la fonction croÃ®t ou dÃ©croÃ®t le plus vite.<br>Nous allons calculer de faÃ§on algorithmique le gradient grÃ¢ce Ã  la Â« diffÃ©rentiation automatique Â».<br> 
<font face="sans-serif" color="#0000FF" size=5ptx style="line-height:25px;">
DÃ©rivÃ©es partielles</font><br>
<font face="sans-serif" color="#00000" size=4ptx style="line-height:25px;">
Pour une fonction de plusieurs variables, il existe une dÃ©rivÃ©e par variable appellÃ©e dÃ©rivÃ©e partielle.<br>
Soit $f:R^2âR$. $\displaystyle \frac{\partial f}{\partial x}(x_0, y_0)$, dÃ©rivÃ©e partielle de $f$ par rapport Ã  $x$ au point $(x_0, y_0) â R^2$ tel que :
    
 - $\displaystyle \frac{\partial f}{\partial x}(x_0, y_0)$, est la dÃ©rivÃ©e en $x_0$ de la fonction d'une variable $xâf(x,y_0)$.
 - $\displaystyle \frac{\partial f}{\partial y}(x_0, y_0)$, est la dÃ©rivÃ©e en $y_0$ de la fonction d'une variable $xâf(x_0,y)$.
    
Autrement dit en revenant Ã  la dÃ©finition de la dÃ©rivÃ©e comme une limite : 
  
$ \displaystyle \frac{\partial f}{\partial x}(x_0, y_0) = \lim\limits_{\substack{h \rightarrow 0 }}\displaystyle \frac{f(x_0+h,y_0)-f(x_0,y_0)}{h}$ et $ \displaystyle \frac{\partial f}{\partial y}(x_0, y_0) = \lim\limits_{\substack{h \rightarrow 0 }}\displaystyle \frac{f(x_0,y_0+h)-f(x_0,y_0)}{h}$<br>

Plus gÃ©nÃ©ralement, pour une fonction $f : R_n â R$ de plusieurs variables: <br> 
 - $\displaystyle \frac{\partial f}{\partial x}(x_1,...,x_n)$ est la dÃ©rivÃ©e partielle de $f$ par rapport Ã  la variable $x_i$ au point $(x_1,...,x_n) â R_n$.    

Câest la dÃ©rivÃ©e en $x_i$ de la fonction dâune variable $x_iâf(x_1,..,x_n)$ considÃ¨rant fixes les variables $x_j$ pour $j \neq i$





</font>

<font face="sans-serif" color="#0000FF" size=6ptx style="line-height:35px;">
<strong>InterprÃ©tation gÃ©omÃ©trique</strong></font><br>
<font face="sans-serif" color="#00000" size=4ptx style="line-height:25px;">
Pour une fonction dâune variable, la dÃ©rivÃ©e est la pente de la tangente au graphe de la fonction (le graphe
Ã©tant alors une courbe). Pour une fonction de deux variables $(x, y) â f (x, y)$, les dÃ©rivÃ©es partielles
indiquent les pentes au graphe de $f$ selon certaines directions (le graphe Ã©tant ici une surface).<br>
    
Plus prÃ©cisÃ©ment :

 - $ \displaystyle \frac{\partial f}{\partial x}(x_0, y_0)$ est la pente du graphe de $f$ en $(x_0, y_0)$ suivant lâaxe $0x$, pente de la tangente Ã  la courbe $z = f (x, y_0)$ donnÃ©e par la dÃ©rivÃ©e $x â f (x, y_0)$ en $x_0$, qui est donc bien $ \displaystyle \frac{\partial f}{\partial x}(x_0, y_0)$â¢   

 - $ \displaystyle \frac{\partial f}{\partial y}(x_0, y_0)$ est la pente du graphe de $f$ en $(x_0, y_0)$ suivant lâaxe $0y$ pente de la tangente Ã  la courbe $z = f (x_0, y)$ donnÃ©e par la dÃ©rivÃ©e $y â f (x_0, y)$ en $y_0$, qui est donc bien $ \displaystyle \frac{\partial f}{\partial y}(x_0, y_0)$â¢

Figure de droite, la dÃ©rivÃ©e partielle $ \displaystyle \frac{\partial f}{\partial y}(x_0, y_0)$indique la pente de la tranche parallÃ¨le Ã  lâaxe $(Ox)$ (en vert).

<strong>Gradient</strong></font><br>
<font face="sans-serif" color=black size=4ptx style="line-height:25px;">
Le gradient est un vecteur dont les coordonnÃ©es sont les dÃ©rivÃ©es partielles. Il a de nombreuses applications
gÃ©omÃ©triques car il donne lâÃ©quation des tangences aux courbes et surfaces de niveau. Surtout, il indique la
direction dans laquelle la fonction varie le plus vite.
    
Soit $f:R^2âR$ admet les dÃ©rivÃ©es partielles le gradient de $f$ en $(x_0, y_0)$ â $R^2$ notÃ© $grad f (x_0,y_0)$, est le vecteur:
    
<br>    
<center>
$grad f (x_0,y_0) =  
\begin{pmatrix}
   \displaystyle \frac{\partial f}{\partial x}(x_0, y_0)  \\
   \displaystyle \frac{\partial f}{\partial y}(x_0, y_0)
\end{pmatrix}$
</center>
<br>    
    Les physiciens et les anglo-saxons notent souvent $\nabla f (x, y)$ pour $grad f(x, y)$. Le symbole $\nabla$ se lit Â« nabla Â».<br>   
Plus gÃ©nÃ©ralement, pour $f : R^n â R$, le gradient de $f$ en $(x_1,. .. , x_n)$ â $R^n$ est le vecteur :<br>
<br>    
<center>
$grad f (x_1,...,x_n) =  
\begin{pmatrix}
   \displaystyle \frac{\partial f}{\partial x_1}(x_1,...,x_n)  \\
   . \\
   \displaystyle \frac{\partial f}{\partial x_n}(x_1,...,x_n)
\end{pmatrix}$
</center>
<br> 
<font face="sans-serif" color=blue size=4ptx style="line-height:25px;">
<i>Remarque : <br>
Le gradient est un Ã©lÃ©ment de $R^n$ Ã©crit comme un vecteur colonne, parfois, pour allÃ©ger lâÃ©criture, on peut Ã©galement lâÃ©crire sous la forme dâun vecteur ligne.</i>   
</font>
</font>


<font face="sans-serif" color=red size=5ptx style="line-height:35px;">
<strong>Tangentes aux lignes de niveau : </strong></font><br>
<font face="sans-serif" color=black size=4ptx style="line-height:25px;">
Soit $f : R^2 â R$ une fonction diffÃ©rentiable. On considÃ¨re les lignes de niveau $f (x, y) = k$.<br><br>
<strong>Proposition 1 :</strong> Le vecteur gradient grad $f (x_0,y_0)$ est orthogonal Ã  la ligne de niveau de $f$ passant au point $(x_0, y_0)$<br><br>
Sur ce premier dessin, sont dessinÃ©s la ligne de niveau passant par le point $(x_0, y_0)$ (en rouge), un vecteur tangent $v$ en ce point et la tangente Ã  la ligne de niveau (en vert).<br><br>
Le vecteur gradient est un vecteur du plan qui est orthogonal Ã  la ligne de niveau en ce point (en bleu).

</div><br>
Ã chaque point du plan, on peut associer un vecteur gradient. Ce vecteur gradient est orthogonal Ã  la ligne
de niveau passant par ce point. Nous verrons juste aprÃ¨s comment savoir sâil est orientÃ© Â« vers le haut Â» ou
Â« vers le bas Â».<br>
Dans le cadre de notre Ã©tude, nous nous intÃ©ressons Ã  lâÃ©quation de la tangente.<br><br>
<strong>Proposition 2 :</strong> Au point $(x_0, y_0)$, lâÃ©quation de la tangente Ã  la ligne de niveau de $f$ est : <br><br>
<center>
$ \displaystyle \frac{\partial f}{\partial x}(x_0, y_0)(x - x_0) { + } \frac{\partial f}{\partial y}(x_0, y_0)(y - y_0) = 0$
